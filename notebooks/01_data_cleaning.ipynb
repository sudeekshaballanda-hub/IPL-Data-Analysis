```python
import pandas as pd
import numpy as np
import os

print("✅ Libraries imported successfully!")
print(f"Pandas version: {pd.__version__}")
```

```python
# Load the datasets
matches = pd.read_csv('../data/matches.csv')
deliveries = pd.read_csv('../data/deliveries.csv')

print(f"✅ Matches dataset: {matches.shape[0]} rows, {matches.shape[1]} columns")
print(f"✅ Deliveries dataset: {deliveries.shape[0]} rows, {deliveries.shape[1]} columns")
print("\n📊 First 5 rows of matches:")
print(matches.head())
```

```python
# Check missing values in matches
print("🔍 Missing values in matches dataset:")
print(matches.isnull().sum())

print("\n🔍 Missing values in deliveries dataset:")
print(deliveries.isnull().sum())
```

```python
# Make a copy before cleaning
matches_clean = matches.copy()

# Drop rows where winner is missing
matches_clean = matches_clean.dropna(subset=['winner'])

# Fix team name inconsistencies
team_name_corrections = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Rising Pune Supergiant': 'Rising Pune Supergiants'
}
matches_clean['winner'] = matches_clean['winner'].replace(team_name_corrections)

print(f"✅ Before cleaning: {matches.shape[0]} rows")
print(f"✅ After cleaning: {matches_clean.shape[0]} rows")
print(f"✅ Removed {matches.shape[0] - matches_clean.shape[0]} rows with missing winners")
```

```python
# Save the cleaned dataset
cleaned_path = '../outputs/cleaned_data/cleaned_matches.csv'
matches_clean.to_csv(cleaned_path, index=False)

print(f"✅ Cleaned data saved to: {cleaned_path}")
print(f"📁 File location: outputs/cleaned_data/cleaned_matches.csv")
```

```python
# Quick statistics to verify
print("📊 Top 5 winning teams:")
print(matches_clean['winner'].value_counts().head())

print("\n📊 Seasons in dataset:")
print(matches_clean['season'].sort_values().unique())
```

```python
# Create a simple summary file
summary = f"""
IPL DATA CLEANING SUMMARY
=========================
Original matches: {matches.shape[0]} rows
Cleaned matches: {matches_clean.shape[0]} rows
Rows removed: {matches.shape[0] - matches_clean.shape[0]}

Top 5 teams by wins:
{matches_clean['winner'].value_counts().head().to_string()}

Seasons covered: {matches_clean['season'].min()} to {matches_clean['season'].max()}
"""

# Save summary
with open('../outputs/insights/data_cleaning_summary.txt', 'w') as f:
    f.write(summary)

print("✅ Summary saved to outputs/insights/")
print("\n" + summary)
```
